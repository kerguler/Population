```python
vec = model("./model.dylib")
```

```python
ret = vec.sim([],
              [100.0, 0.0, 20.0, 5.0, 5.0, 1.0, 10.0], 
              40,
              10)
```

```python
ret.shape
```

```python
retp = numpy.percentile(ret,[5,50,95],axis=0)
```

```python
xr = numpy.arange(retp.shape[1])

plt.fill_between(xr, retp[0,:,0], retp[2,:,0], alpha=0.5)
plt.plot(xr, retp[1,:,0])
plt.show()

plt.fill_between(xr, retp[0,:,1], retp[2,:,1], alpha=0.5)
plt.plot(xr, retp[1,:,1],'o-')
plt.show()
```
